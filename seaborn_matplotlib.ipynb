## convinando estilos

```python
#datos1=np.random.uniform(-10,10,size=1000)
datos1=np.random.randn(10000)
datos2 = np.random.normal(0, 1, 10000)
```

```python
sns.displot(datos1, kde=True, bins=50)
```

```python
sns.displot(datos2,kde=True, bins=75)
```

```python

sns.kdeplot(datos2, color="green", linewidth=2)
#mpl.xlabel('Valores')
#mpl.ylabel('Densidad')
#mpl.title('Distribución de datos')
#mpl.show()
```

```python
argumentos_curva = {'color':'blue','label':'curva'}
argumentos_histograma = {'color':'black','label':'histograma'}
sns.displot(datos2, bins=25, kde_kws=argumentos_curva, hist_kws=argumentos_histograma)
```

```python
import matplotlib.pyplot as plt



sns.histplot(datos2, bins=25, color='black', label='histograma')
sns.kdeplot(datos2, color='blue', label='curva')
plt.xlabel('Valores')
plt.ylabel('Frecuencia / Densidad')
plt.title('Distribución de datos')
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt


sns.displot(datos2, bins=25, kde=True, color='blue', label='Datos')
sns.displot(datos2, bins=25, kde=True, color='red', label='Curva de densidad', common_norm=False)

plt.xlabel('Valores')
plt.ylabel('Frecuencia / Densidad')
plt.title('Distribución de datos')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt


sns.histplot(datos2, bins=25, color='gray', label='Histograma')
sns.kdeplot(datos2, color='green', label='Curva de densidad')

plt.xlabel('Valores')
plt.ylabel('Frecuencia / Densidad')
plt.title('Distribución de datos')
plt.legend()
plt.show()
```

# Tratamiento de datos con DataFrames

## Unión de DataFrames

```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.DataFrame({'Col1':['1','2','3'], 'clave':['a','b','c']})
df_1
```

```python
df_2 = pd.DataFrame({'col2':['4','5','6'], 'clave':['c','b','e']})
df_2
```

```python
df_3=pd.DataFrame.merge(df_1,df_2)
df_3
```

```python
df_3=pd.DataFrame.merge(df_1,df_2, on='clave')
df_3
```

```python
df_4 = pd.DataFrame.merge(df_1,df_2, on='clave', how='left')
df_4
```

```python
df_5 = pd.DataFrame.merge(df_1,df_2, on='clave', how='right')
df_5
```

```python
df_6 = pd.DataFrame.merge(df_1,df_2, on='clave', how='outer')
df_6
```

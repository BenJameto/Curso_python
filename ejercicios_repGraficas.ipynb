```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
enteros=np.random.randint(0,500,100)
matriz_enteros=enteros.reshape(10,10)
sns.heatmap(matriz_enteros, annot=True, fmt='d', cmap='YlOrRd')
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.show()
```

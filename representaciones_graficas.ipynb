# Diagrama de caja

### sirve para representar graficamente una serie de datos numericos a traves de sus cuartiles

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datos = np.random.randn(100)
```

```python
sns.boxplot(datos)
```

```python
sns.kdeplot(datos, color = 'green')
sns.displot(datos, kde=True, bins=100)
```

# Regresiones

```python
propinas = sns.load_dataset('tips')
```

```python
propinas.head(20)
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas)
```

```python
sns.lmplot(x='total_bill', y='tip', data= propinas, scatter_kws={'marker':'o', 'color':'green'}, line_kws={'color':'blue'})
```

```python
sns.lmplot(x='total_bill', y='tip', data= propinas, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
sns.lmplot(x='total_bill', y='tip', data= propinas, scatter_kws={'color':'green'}, fit_reg=False)
```

```python
propinas['porcentaje']= 100 * propinas['tip'] / propinas['total_bill']
```

```python
propinas
```

```python
sns.lmplot(x='size', y='porcentaje', data=propinas)
```

```python
sns.lmplot(x='total_bill', y='porcentaje', data=propinas, hue='sex', markers=['x','o'])
```

```python
sns.lmplot(x='total_bill', y='porcentaje', data= propinas, hue='day')
```

# Mapas de calor (heatmap)

```python
vuelos = sns.load_dataset('flights')
```

```python
vuelos
```

```python
vuelos = vuelos.pivot(index='month', columns= 'year', values= 'passengers')
```

```python
vuelos
```

```python
vuelos.index
```

```python
sns.heatmap(vuelos)
```

```python
sns.heatmap(vuelos, annot=True, fmt='d')
```

```python
valor_central=vuelos.loc['May'][1956]
```

```python
sns.heatmap(vuelos, center=valor_central, annot=True, fmt='d')
```

```python
sns.heatmap(vuelos, center=valor_central, annot=True, fmt='d', cbar_kws={'orientation':'horizontal'})
```

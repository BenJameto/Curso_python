# Diagrama de caja

### sirve para representar graficamente una serie de datos numericos a traves de sus cuartiles

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datos = np.random.randn(100)
```

```python
sns.boxplot(datos)
```

```python
sns.kdeplot(datos, color = 'green')
sns.displot(datos, kde=True, bins=100)
```
